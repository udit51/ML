```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loaded datasets in seaborn
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
trains = pd.read_csv('titanic.csv')
```

```python
trains.head()
```

```python
#BIvaraite Analysis
''' Scatterplot  NUMERICAL NUMERICAL ANALYSIS'''
sns.scatterplot(x='total_bill', y='tip', data=tips) # Pass data and column names as keyword arguments
#sns.scatterplot(tips['total_bill'], tips['tip'])   error
```

```python
#multivariate on same data -> ek saath analysis on 2 or more columns
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker' ,size = 'size', data=tips)

#here we are getting to study on 5 points on a dataset
```

```python
#Bar plot
''' NUMERICAL - CATEGORICAL DATA '''
#trains.head()
sns.barplot(x='Pclass', y='Fare', hue='Sex', data=trains)
#the black line on every bar tells the Confidence Interval ....
```

```python
#Distplot  -> basically gives the probability dnsity function
sns.distplot(trains[trains['Survived']==0]['Age'],hist=False)  # survived == 0 means jo mar gye woh aagye
sns.distplot(trains[trains['Survived']==1]['Age'],hist=False)  #jo survive kr gye
```

```python
#HEAT MAP
#Categorical - categorical analysis
#we can use cross tab function  -> gives the frequency distribution of the data
pd.crosstab(trains['Pclass'],trains['Survived'])

```

```python
sns.heatmap(pd.crosstab(trains['Pclass'],trains['Survived']))
```

```python
trains.groupby('Embarked')['Survived'].mean()*100
#or
# Convert 'Survived' column to numeric before calculating the mean if required
# trains['Survived'] = pd.to_numeric(trains['Survived'], errors='coerce')  # errors='coerce' converts invalid values to NaN
#trains.groupby('Pclass').mean()['Survived']*100 '''
```

```python
#CLUSTER MAP
sns.clustermap(pd.crosstab(trains['SibSp'], trains['Survived'])) # Pass 'Survived' as the 'columns' argument to pd.crosstab()
```
